```python
import numpy as np

import matplotlib.pyplot as plt

for damping in [0.1, 0.2, 0.3, 0.5, 0.6, 0.7, 0.9]:
    n = np.arange(0, 11)
    y = np.exp(-damping*n)
    plt.plot(n, y, label=f'damping = {damping}')

plt.xlabel('n')
plt.grid(True)
plt.legend()
plt.xlim(0, 10)
plt.ylim(0, 1)
plt.show()
```
